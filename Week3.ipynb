```
# Let's import the pandas library
#

import pandas as pd
# as well as the Numpy library
import numpy as np
```

```
#Telecom dataset 1
with open('/content/Telcom_dataset.csv','r') as f:
    df = pd.read_csv(f, header=1)
  
df.to_csv('dataset1.csv')

# Loading the csv
df = pd.read_csv('dataset1.csv')
df
df = pd.read_csv('dataset1.csv', header=1) 
df = pd.read_csv('dataset1.csv', names=['UID', 'PRODUCT', 'VALUE',  'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A','COUNTRY_B', 'CELL_ID', 'SITE_ID' ])
#Drop SITE_ID because it is similar to CELL_ID
df.drop('SITE_ID', axis=1, inplace=True)
df



```

```
with open('/content/Telcom_dataset2.csv','r') as f:
    df2 = pd.read_csv(f, header=1)
  
df2.to_csv('dataset2.csv')

# Loading a csv
df2 = pd.read_csv('dataset2.csv')
df2
df2 = pd.read_csv('dataset2.csv', header=1) 
df2 = pd.read_csv('dataset2.csv', names=['UID', 'PRODUCT', 'VALUE',  'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A','COUNTRY_B', 'CELL_ID', 'SITE_ID' ])
#Drop SITE_ID because it is similar to CELL_ID
df2.drop('SITE_ID', axis=1, inplace=True)
df2
```

```
#Telecom dataset 3
with open('/content/Telcom_dataset3.csv','r') as f:
   df3 = pd.read_csv(f, header=1)
df3.to_csv('dataset3.csv')
# Loading the csv
df3 = pd.read_csv('dataset3.csv')
df3
df3 = pd.read_csv('dataset3.csv', header=1) 
df3 = pd.read_csv('dataset3.csv', names=['UID', 'PRODUCT', 'VALUE',  'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A','COUNTRY_B', 'CELL_ID', 'SITE_ID' ])
#Drop SITE_ID because it is similar to CELL_ID
df3.drop('SITE_ID', axis=1, inplace=True)
df3
```

```
#Merging The three databases
df_all = pd.concat((df, df2, df3), axis=0, ignore_index=True)
df_all
```

```
#cleaning of the data 

df_all.isnull()
df_no_missing = df_all.dropna()
df_no_missing
df_cleaned = df_all.dropna(how='all')
df_cleaned
df_all.fillna(0)
df_all.drop('UID', axis= 1)

```

```
#Loading our cell geo database
with open('/content/cells_geo.csv','r') as f:
    df_geo = pd.read_csv(f, header=1)
  
df_geo.to_csv('cellgeo.csv')

# Loading the csv
df_geo = pd.read_csv('cellgeo.csv')
df_geo
df_geo = pd.read_csv('cellgeo.csv', header=1) 
df_geo = pd.read_csv('cellgeo.csv', names=['UID', 'CITY', 'STATUS',  'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE','REGION', 'AREA', 'CELL_ID', 'SITE_CODE' ])

df_geo
```

```

#Clean the cellgeo database 
df_geo.dropna()
df_geo.drop('UID', axis= 1)
```

```
#rename all null values in area to unnamed.Basically we still cleaning our data 
df_city = df_geo.replace([np.NaN], 'Unnamed: 9')
df_city
```

```
col_to_use = df_city.columns.difference(df_all.columns)
df_combine = pd.concat([df_all, df_city[col_to_use]], join='inner', axis=1)
df_combine  # concats both databases and removes one CELL_ID column as it will be  a duplicate on merged cells 

```

```
#Cleans the new database values
df_combine.isnull()
df_no_miss = df_combine.dropna()
df_no_miss
df_cleaned = df_combine.dropna(how='all')
df_cleaned
df_combine.fillna(0)
df_combine.drop('UID', axis= 1)
```

```
 #Most used city in the three days 
 df_combine['CITY'].max()
```

```
#Least used city in three days 
df_combine['CITY'].min()
```

```
#Change DATE_TIME to datetime datatype
df_combine.dtypes
df_combine['DATA_TIME']= pd.to_timedelta(df_combine['DATE_TIME'], errors='coerce')
```

```
#Which cities  used most expenditure  during business  hours?

df_time= df_combine[(df_combine['DATE_TIME'] > '08.00.00' ) & (df_combine['DATE_TIME'] < '17.00.00')]
list(set(df_time['CITY']))
grouped = df_time.groupby(['CITY', 'PRODUCT'])['VALUE'].max()
grouped
```

```
#Which cities were the most expenditure  used during  home hours?
df_time1= df_combine[ (df_combine['DATE_TIME'] < '05.00.00') ]
df_time2 = df_combine[(df_combine['DATE_TIME'] > '17.00.00') ]
both = pd.concat([df_time1, df_time2])
group = both.groupby(['CITY', 'PRODUCT'])['VALUE'].max()
group
```
